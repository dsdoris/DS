## 아래 만들어진 함수를 이용하여 XOR GATE를 제작해보세요. 

```
def AND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = x1*w1 + x2*w2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1

def  NAND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = x1*w1 + x2*w2
  if tmp <= theta:
    return 1 ## only change here
  elif tmp > theta:
    return 0 ## only change here

def OR(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.3 ## only change here
  tmp = x1*w1 + x2*w2
  if tmp <= theta:
    return 0 
  elif tmp > theta:
    return 1 
```

이 게이트는 두 입력 값 x1, x2가 다를 때만 1을 출력합니다.

(Hint: 위에서 만든 함수를 종합해서(묶어서) 흐름을 만들도록 활용하면 됩니다)

XOR(1, 1) = 0

XOR(1, 0) = 1

XOR(0, 1) = 1

XOR(0, 0) = 0

## 신경망에서 용어들에 대해서 자신만의 언어로 설명해보세요. 

인터넷 자료를 찾아도 좋으나, 그대로 쓰지 마시고, 내 언어로 바꾸어 설명해보시는 것이 중요합니다.

먼저 영문 이름을 한글로 바꿔보시고, 영문과 한글이 동시에 머리속에 들어오면 이 과제는 성공입니다. 

**Input Layer(입력 층):** 데이터를 입력받는 부분으로, 받은 데이터를 hidden node로 전달한다.

**Hidden Layer(은닉 층):** 전달받은 데이터를 가지고 계산하는 부분, 계산 결과가 보이지 않아서 hidden이라는 이름이 붙었다.

**Output Layer(출력 층):** 계산된 결과 값을 출력하는 부분

**Neuron(신경 세포):** 사람의 뇌에서 뉴런이라는 신경세포를 통해 신경 전달물질을 전달하면서 정보를 처리하는데, 컴퓨터의 인공지능 학습에 이 방법을 적용하여 '인공 신경망(Artificial Neurol Network)'이라는 딥러닝이 생겼다. 인공신경망 기법에서는 각 node들이 서로 정보를 전달, 계산, 전달하면서 데이터를 학습하고 결과를 낸다.

**Weight(가중치):** Neural network에서 어떤 값을 예측하는 데 있어서 얼마나 중요한 영향을 미치는지의 정도를 말한다. 어떤 뉴런(노드)에서 계산된 값에 곱해주기 때문에, 계산된 값이 이 가중치에 따라서 달라지기도 한다. => 입력에 맞는 출력이 나올 수 있도록 가중치를 조절해야 한다.

**Bias(편향):** 예측하려는 특성 자체가 가진 성향(편향)을 말한다. 같은 가중치를 가졌더라도 특성 자체가 한쪽으로 치우쳐진 성향이 있다면 예측 결과가 달라진다.

**Activation Function(활성 함수):** 출력 값을 0 또는 1 (이진값)으로 바꿔주는 함수를 말한다. 

**Node Map(노드 지도):** 인공신경망 기법에서 각 노드들의 상호작용 관계를 표시한 그림을 말한다.

**Perceptron:** 뉴런이 정보를 입력받고 출력하는 것을 모방해서 만든 인공지능 알고리즘이다. 입력은 여러 개가 될 수 있으나 한 개의 값을 출력한다.

**Epoch:** 전체 트레이닝 셋이 신경망을 통과한 횟수를 의미한다.
